```python
from matplotlib import pyplot as plt
import scipy as sp
from matplotlib import cm
```

```python
start = sp.zeros((100, 100))
end = sp.zeros((100, 100))
with open("start_end_grids.txt") as f:
    current = start
    # skip the 100x100 line
    next(f)
    for line in f:
        if line.strip() == "100x100":
            current = end
            continue
        line = line.strip().split()
        row = int(line[0].strip(':')) - 1
        for (col, value) in enumerate(line[1:]):
            current[row, col] = float(value)
```

```python
start_fig = plt.figure()
start_axe = start_fig.add_subplot(1, 1, 1)
start_axe.imshow(start, cmap=cm.coolwarm, origin='lower')
```

```python
end_fig = plt.figure()
end_axe = end_fig.add_subplot(1, 1, 1)
end_axe.imshow(end, cmap=cm.coolwarm, origin='lower')
```
